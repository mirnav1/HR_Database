```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import sqlite3
from sqlite3 import Error
import pandas as pd

con = sqlite3.connect('HR.db')
cursor = con.cursor()
```

```python
def create_connection(db_file):
	conn = None
	try:
		conn = sqlite3.connect(db_file)
		print(sqlite3.version)
	except Error as error:
		print(error)
	finally:
		if conn:
		    conn.close()
if __name__ == '__main__':
		create_connection(r"HR.db")
```

Retrieve all employees whose address is in Elgin,IL.

```python
df_Elgin = pd.read_sql_query("SELECT F_NAME , L_NAME FROM EMPLOYEES WHERE ADDRESS LIKE '%Elgin,IL%'", con)
df_Elgin
```

Retrieve all employees who were born during the 1970's.

```python
df_1970 = pd.read_sql_query("SELECT F_NAME, L_NAME FROM EMPLOYEES WHERE B_DATE LIKE '%197%';", con)
df_1970
```

Retrieve all employees in department 5 whose salary is between 60000 and 70000.

```python
df_SALARY_DEP_5 = pd.read_sql_query("SELECT * FROM EMPLOYEES WHERE DEP_ID = 5 AND (SALARY BETWEEN 60000 AND 70000)", con)
df_SALARY_DEP_5
```

Retrieve a list of employees ordered by department ID.

```python
df_EMPLOYEES_DEP = pd.read_sql_query("SELECT F_NAME,L_NAME,DEP_ID FROM EMPLOYEES ORDER BY DEP_ID ASC;", con)
df_EMPLOYEES_DEP
```

Retrieve a list of employees ordered in descending order by department ID and within each department ordered (alphabetically) in descending order by last name.

```python
df_EMPLOYEES_DEP_ID_DESC = pd.read_sql_query("SELECT F_NAME,L_NAME,DEP_ID FROM EMPLOYEES ORDER BY DEP_ID DESC, L_NAME DESC;", con)
df_EMPLOYEES_DEP_ID_DESC
```

Retrieve a list of employees ordered by department name and within each department ordered (alphabetically) in descending order by last name.

```python
df_EMPLOYEES_DEP_NAME_DESC = pd.read_sql_query("SELECT E.F_NAME, E.L_NAME, D.DEP_NAME FROM EMPLOYEES AS E, DEPARTMENTS AS D WHERE E.DEP_ID = D.DEPT_ID_DEP ORDER BY D.DEP_NAME, L_NAME DESC;", con)
df_EMPLOYEES_DEP_NAME_DESC
```

For each department retrieve the number of employees in the department and the average employee salary in the department. Order the result set by Average Salary.

```python
df_EMPLOYEES_DEP_SALARY = pd.read_sql_query("SELECT DEP_ID, COUNT(*), AVG(SALARY) FROM EMPLOYEES GROUP BY DEP_ID HAVING COUNT(*) ORDER BY AVG(SALARY) ASC;", con)
df_EMPLOYEES_DEP_SALARY
```

Select the names, job start dates, and job titles of all employees who work for the department number 5.

```python
df_EMPLOYEES_DEP_5 = pd.read_sql_query("SELECT E.F_NAME,E.L_NAME,JH.START_DATE, J.JOB_TITLE FROM EMPLOYEES AS E INNER JOIN JOB_HISTORY AS JH ON E.EMP_ID=JH.EMPL_ID INNER JOIN JOBS AS J ON E.JOB_ID=J.JOB_IDENT WHERE E.DEP_ID ='5';", con)
df_EMPLOYEES_DEP_5
```

Select employee id, last name, department id and department name for all employees.

```python
df_EMPLOYEES_DEP_ALL = pd.read_sql_query("SELECT E.EMP_ID, E.L_NAME, E.DEP_ID, D.DEP_NAME FROM EMPLOYEES AS E LEFT OUTER JOIN DEPARTMENTS AS D ON E.DEP_ID=D.DEPT_ID_DEP;", con)
df_EMPLOYEES_DEP_ALL
```
